```python
# Kiểm tra một vài dòng transcript
import pandas as pd

metadata = pd.read_csv("LJSpeech-1.1/metadata.csv", sep='|', header=None)
metadata.columns = ['filename', 'transcript', 'normalized']
metadata.head()
```

```python
from sklearn.model_selection import train_test_split

train, temp = train_test_split(metadata, test_size=0.4, random_state=42)
val, test = train_test_split(temp, test_size=0.5, random_state=42)

print(f"Train: {len(train)}, Val: {len(val)}, Test: {len(test)}")
```

```python
import torch

# Tải Tacotron 2 pretrained
tacotron2 = torch.hub.load('nvidia/DeepLearningExamples:torchhub', 'nvidia_tacotron2')
tacotron2 = tacotron2.eval().cuda()

# Tải WaveGlow pretrained
waveglow = torch.hub.load('nvidia/DeepLearningExamples:torchhub', 'nvidia_waveglow')
waveglow = waveglow.remove_weightnorm(waveglow).eval().cuda()

# Tắt gradient
for m in [tacotron2, waveglow]:
    for param in m.parameters():
        param.requires_grad = False
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import write
from IPython.display import Audio

# Tiền xử lý văn bản
from tacotron2.text import text_to_sequence

def tts(text, tacotron2, waveglow):
    # Chuẩn bị input sequence
    sequence = np.array(text_to_sequence(text, ['english_cleaners']))[None, :]
    sequence = torch.from_numpy(sequence).cuda().long()
    input_lengths = torch.LongTensor([sequence.size(1)]).cuda()

    # Tacotron2: text -> mel spectrogram
    with torch.no_grad():
        mel_outputs = tacotron2.infer(sequence, input_lengths)[0]

    # WaveGlow: mel -> audio waveform
    with torch.no_grad():
        audio = waveglow.infer(mel_outputs)
        audio = audio[0].data.cpu().numpy()

    return audio

```

```python

# Dùng thử
text = "Hello"
audio = tts(text, tacotron2, waveglow)

# Nghe thử
Audio(audio, rate=22050)
```
